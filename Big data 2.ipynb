```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
covid_data = pd.read_csv("Effects-of-COVID-19-on-trade-1-February-18-November-2020-provisional.csv")
```

```python
covid_data
```

```python
plt.bar(covid_data.Direction, covid_data.Cumulative, align = "center")
plt.rcParams["figure.figsize"] = [5,5]
```

```python
imp = covid_data[covid_data.Direction == "Imports"]
exp = covid_data[covid_data.Direction == "Exports"]
reimp = covid_data[covid_data.Direction == "Reimports"]
```

```python
plt.plot(imp.Country, imp.Value, ".")
plt.legend(["Country"])
plt.xlabel("Country")
plt.ylabel("Value")
plt.title("Importation")
plt.show()
```

```python
plt.plot(exp.Country, exp.Value, ".")
plt.legend(["Country"])
plt.xticks(rotation = 45)
plt.xlabel("Country")
plt.ylabel("Value")
plt.rcParams['figure.figsize'] = [20,20]
plt.title("Exportation")
plt.show()
```

```python
plt.plot(reimp.Country, reimp.Value, "red")
plt.legend(["Country"])
plt.xlabel("Country")
plt.ylabel("Value")
plt.title("Reimportation")
plt.show()
```

```python
plt.plot(imp.Year, imp.Value, "*")
plt.legend(["Year"])
plt.xlabel("Year")
plt.ylabel("Value")
plt.title("Importation")
plt.show()
```

```python
plt.plot(exp.Year, exp.Value, "*")
plt.legend(["Year"])
plt.xlabel("Year")
plt.ylabel("Value")
plt.title("Exportation")
plt.show()
```

```python
plt.plot(reimp.Year, reimp.Value, "*")
plt.legend(["Year"])
plt.xlabel("Year")
plt.ylabel("Value")
plt.title("Reimportation")
plt.show()
```

```python
plt.plot(imp.Transport_Mode, imp.Value, "o")
plt.legend(["Transport Mode"])
plt.xlabel("Transport Mode")
plt.ylabel("Value")
plt.rcParams["figure.figsize"] = [5, 5]
plt.title("Importation")
plt.show()
```

```python
plt.plot(exp.Transport_Mode, exp.Value, "o")
plt.legend(["Transport Mode"])
plt.xlabel("Transport Mode")
plt.ylabel("Value")
plt.rcParams["figure.figsize"] = [10, 5]
plt.title("Exportation")
plt.show()
```

```python
plt.plot(reimp.Transport_Mode, reimp.Value, "o")
plt.legend(["Transport Mode"])
plt.xlabel("Transport Mode")
plt.ylabel("Value")
plt.rcParams["figure.figsize"] = [10, 5]
plt.title("Reimportation")
plt.show()
```

```python
plt.plot(imp.Commodity, imp.Value, "+")
plt.legend(["Commodity"])
plt.xlabel("Commodity")
plt.ylabel("Value")
plt.title("Importation")
plt.show()
```

```python
plt.plot(exp.Commodity, exp.Value, "+")
plt.xticks(rotation = 45)
plt.legend(["Commodity"])
plt.xlabel("Commodity")
plt.ylabel("Value")
plt.title("Exportation")
plt.show()
```

```python
plt.plot(reimp.Commodity, reimp.Value, "+")
plt.legend(["Commodity"])
plt.xlabel("Commodity")
plt.ylabel("Value")
plt.title("Reimportation")
plt.show()
```
